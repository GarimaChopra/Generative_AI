```python

text= 'The black cat sat on the couch and the brown dog slept on the rug'
# tokenize the sentence into words
tokenized_text= text.lower().split()

print(tokenized_text)
```

```python
from gensim.models import Word2Vec
sentence=[tokenized_text]
print(sentence)
model= Word2Vec([tokenized_text], min_count=1,vector_size=512)
print(model.wv.get_vector('dog'))
```

```python
print("Cosine similarity between 'black' " +
          "and 'brown' ",
    model.wv.similarity('cat', 'dog'))

```

```python
print("Cosine similarity between 'couch' " +
            "and 'rug'  ",
      model.wv.similarity('couch', 'rug'))
```

```python
print("Cosine similarity between 'cat' " +
          "and 'couch' ",
    model.wv.similarity('cat', 'couch'))
```
